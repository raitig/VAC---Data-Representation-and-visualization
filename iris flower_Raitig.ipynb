```python
df=np.arange(10,15)
df1=np.array([11,12,13,14,15])
print(df)
print(df1)
np.corrcoef(df,df1)
```

```python
df=np.arange(25,30)
df1=np.array([11,15,23,25,29])
r=np.corrcoef(df,df1)
r1=scipy.stats.pearsonr(df,df1)
print(r1[0],r1[1])
r2=scipy.stats.spearmanr(df,df1)
print(r2[0],r2[1])
r3=scipy.stats.kendalltau(df,df1)
print([r3[0],r3[1]])
```

```python
df=np.arange(25,30)
df1=np.array([11,15,23,25,29])
r=np.corrcoef(df,df1)
r1,p1=scipy.stats.pearsonr(df,df1)
print(r1,p1)
r2,p2=scipy.stats.spearmanr(df,df1)
print(r2,p2)
r3,p3=scipy.stats.kendalltau(df,df1)
print(r3,p3)
```

```python
reg=scipy.stats.linregress(df,df1)
print(reg)
print(reg.slope)
print(reg.rvalue)
print(reg.pvalue)
print(reg.stderr)
print(reg.intercept)
print(reg.intercept_stderr)
```
